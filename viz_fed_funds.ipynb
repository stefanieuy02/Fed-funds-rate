```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
int_rate = pd.read_csv('FEDFUNDS.csv') #from FRED stlouis website!
```

```python
int_rate.info()
```

```python
int_rate.head(10)
```

```python
x = list(int_rate['DATE'])
y = list(int_rate['FEDFUNDS'])
label = list(int_rate['DATE'])

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (20,6))

plt.grid()
plt.plot(x,y)
ax.set_xticks(label[::25]) #frequency of x tick labels 
ax.set_xticklabels(label[::25], rotation=45)
ax.set_title('Federal Funds Rate from 1954-2021')
ax.set_xlabel('Date')
ax.set_ylabel('Fed Funds Rate')
```
